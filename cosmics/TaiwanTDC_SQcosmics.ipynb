## Introduction to timing/trigger information:

The current cosmic data of SpinQuest contains only basic infomation of the signals received by each of the detectors.
The analog signals of the energy loss of charged particles in the detectors are digitized with time-to-digital converter (TDC) modules.

### TDC time
The current dataset that you will be using uses only signals from the so-called `Taiwan TDCs`, to do some rough-alignment in the signal timing.
A different alignment needs to be performed to the main and to the Dark Photon trigger system.

One piece of information saved in the following data files is the TDC Time. 
The TDC time of a hit is the time difference between its arrival time and the stop signal. 
In the current Data adquisition system (DAQ) there is a common stop signal issued by the trigger supervisor in a certain time window.
The TDC time is the amount of time between the hit arrival and the common stop signal. 

Ideally we want this difference in time as similar as possible for all the recorded events.
Therefore if one were to plot the distribution of tdcTime for all hits we would expect to have a sharp peak.
Sometimes a track will produce multiple hits within the same pre-defined period of time, producing a second peak in the tdc distribution. 
This could also come from random or non-physical tracks.
It would be nice if in the following we could sort out good cosmic muons from random tracks.

In general we want to analyze the hit patter on the hodoscope and dark photon hodoscope planes to identify candidate muon tracks.

### Triggers

The triggers that this data uses are NIM triggers. 
Since NIM cosmic triggers are random with respect to the RF clock, the TDC time distributions should be wider than
“without RF integration condition”.
Even though the TDC timing resolution could be close to 1ns, we won't expect to get that alignment with cosmics.
The fine alignment will come once we have beam commisioning data and the channels can be re-alignemd with the beam RF (Radio frequency) clock.

The list of the current triggers in the MainDAQ is:

- H1&&H2&&H3&&H4: (H1X||H1Y) && (H2X||H2Y) && (H3X||H4Y1)&& (H4X||H4Y2)
- H1&&H2 : (H1X||H1Y) && (H2X||H2Y)
- H2&&H4 : (H2X||H2Y) && (H4X||H4Y2)

```python
# open and convert to panda df
runNum = '001573'
hits = uproot.open("data/run_%s.root"%runNum)["save"]
hitsdf = hits.pandas.df()
hitsdf.columns
```

Each row corresponds to one hit.
Each hit has the following properties:

1. __rocID__: detector that received that hit
    For our purposes:
                DP1: rocID==31 
                DP2: rocID==13
2. __boardID__: TDC boards, each detector is connected to a number of TDC boards - this number can be used to map to each quadrant. Each crate has up to 7 TDC ports, from left to right.
    
3. __channelID__: channel number of the TDC: each TDC has 64 channels so this goes from 0 - 64
   Eventually we need to convert this to the channel number of the detector itself (we call this 'elementID').
   For example: 
                DP1: has 80 channels per quadrant
                DP2: has 50 channels per quadrant

4. __eventID__: The event number. For a certain event number there will be multiple hits. Ideally if you want to see hits in the same event you may want to reorganize your pandas dataframe by the same eventID.

5. __tdcTime__: The amount of time between the hit arrival and the common stop signal, this is in nanoseconds.

6. __eventTy__: The ID of the NIM trigger triggered by this hit. 

The full mapping file can be found in `param_2020012301.tsv'.

```python
# read columns
tdcTime = hits.array("tdcTime").flatten()
channelID = hits.array("channelID").flatten()
boardID = hits.array("boardID").flatten()
rocID = hits.array("rocID").flatten()
eventTy = hits.array("eventTy").flatten()
```

Now we first want to build masks to identify hits in the DP1 and DP2 detectors.

```python
# mask rocID for DP hodoscopes
maskDP1 = (rocID==31)
maskDP2 = (rocID==13)
```

Now we want to be more specific about the position of the hit and identify this by quadrant.
We can use the boardID to map the quadrant easily.

For DP2 we have 3 TDC boards.
For DP1 we have 1 TDC board.
BoardID is saved in 0-6 in the ntuple, but in the mapping file we can see that is saved as an hexadecimal number:

    det_name ele roc board  chan
    DP1TL    1   31  2304.  0

It's a 4 hexadecimal number, the first boardID 0 -> 0x0900, this is 2304 in decimal number.

    boardID = 0, 0x0900=2304,
    boardID = 1, 0x0a00=2560,
    boardID = 2, 0x0b00=2816,
    boardID = 3, 0x0c00=3072,
    boardID = 4, 0x0d00=3328,
    boardID = 5, 0x0e00=3584,
    boardID = 6, 0x0f00=3840,

We also do the last mapping from the TDC channel to the detector channel (also called elementID).
Again we use the mapping file for this. 

This part is a bit annoying because the channelID only goes up to 63 while the detectors could have more or overlappping number of channels.

So e.g. for the same quadrant e.g. `DP2TR` you have 2 types of boards connected (0 and 1) and the TDC chaneels (48-63) and (0-31), you want to map these to the elementID channels 0-50. 

```python
# map of boardID for quadrant in DP hodoscopes
boards = {'DP2TL': [[0,0,47]],
          'DP2TR': [[0,48,63],[1,0,31]],
          'DP2BL': [[1,32,63],[2,0,15]],
          'DP2BR': [[2,16,63]],
          'DP1TL': [[0,0,63],[1,0,15]],
          'DP1TR': [[1,16,63],[2,0,31]],
          'DP1BL': [[2,32,63],[3,0,47]],
          'DP1BR': [[3,48,63],[4,0,63]],
          }          
```

```python
# build masks of boardID and rocID for DP
masks = {}
masksToID = {}
for key,it in boards.items():
    mask = False
    masksToID[key] = {}
    for ib,b in enumerate(it):
        mask_b = np.logical_and.reduce([(boardID==b[0]), (channelID>=b[1]), (channelID<=b[2])])
        mask = np.logical_or(mask, mask_b)
        # mask, number to subtract, number to add
        tosubtract = 0
        if b[1] > 0:
            tosubtract = b[1] 
        toadd = 0
        if ib> 0:
            toadd = it[ib-1][2]-it[ib-1][1]+1
        masksToID[key]['%i'%b[0]] = [mask_b, tosubtract,toadd]
    masks[key] = mask
```

```python
# now change channelID to elementID (i.e. which bar of DP has hits)
elementID = channelID.copy()
for key,dicti in masksToID.items():
    for k,m in dicti.items():
        #print('for %s and board %s, subtract %i and add %i'%(key, k, m[1],m[2]))
        if 'DP1' in key:
            mDP = maskDP1
        else:
            mDP = maskDP2
        mask = np.logical_and(m[0],mDP)
        elementID[mask] = elementID[mask] - m[1] + m[2]
```

Now we build the trigger masks. We can take up to 10 different triggers and you consider this a 10bit binary number. 
I.e. if first trigger is fired then first bit is set to 1000000000.

In this case the mapping for the triggers we are interested on is the following:

   - __NIM2__: H1 && H2: 0b1000000 = 0x40 = 64 i.e. if eventTy==64 => random trigger on H1&&H2

   - __NIM4__: H2 && H4: 0b100000000 = 0x100 = 256 i.e. if eventTy==256 => random trigger on H2&&H4  

```python
# use H1&&H2 to trigger on DPs
triggerDP1 = (eventTy==64)
triggerDP2 = (eventTy==64)
#triggerDP2 = (eventTy==256)
```

```python
# now implement all masks and plot TDC time as function of elementID
for key,it in boards.items():
    # labels and masks
    if 'DP1' in key:
        mDP = maskDP1
        binsDP = 80-1
        mtrigger = triggerDP1
        triggerlab = 'H1 && H2'
    else:
        mDP = maskDP2
        binsDP = 50-1
        mtrigger = triggerDP2
        triggerlab = 'H1 && H2'
        #triggerlab = 'H2 && H4'
    mask = np.logical_and.reduce([masks[key],mDP,mtrigger])

    # plot hits per elementID
    plt.clf()
    plt.hist(elementID[mask], bins=binsDP, range=(0, binsDP))
    plt.xlabel('Channel %s'%key)
    plt.ylabel('Number of hits / channel')
    plt.title('After %s'%triggerlab)
    plt.show()
    plt.clf()

    # plot tdcTime for all hits
    plt.hist(tdcTime[mask], bins=40, range=(800, 1100))
    plt.xlabel('TDC time %s'%key)
    plt.ylabel('Number of hits / channel')
    plt.title('After %s'%triggerlab)
    plt.show()
    plt.clf()

    # 2d plot of tdcTime vs elementID
    fig, ax = plt.subplots()
    counts, xedges, yedges, im = ax.hist2d(elementID[mask], tdcTime[mask], bins=(binsDP,40))
    plt.ylabel('TDC time %s'%key)
    plt.xlabel('Channel %s'%key)
    plt.title('After %s'%triggerlab)
    plt.colorbar(im, ax=ax)
    plt.show()
    plt.clf()
```

The following is a function fo map a channelID to a position. Since the detectors that we will be mostly working with are scintillator hodoscopes, we mostly have an idea of the position in (y) and where the detector is placed w.r.t. the beam axis.
For now only DP detectors are included. You can find the other detector information in the literature, e.g.:
<img src="data/table-hodos.png" width=1000 height=1000 />

```python
# for now only DP information
def mapChannelToPosition(channel, detector):
    chan = channel
    det = detector

    # bar widhts in cm
    barwidhts = {}
    barwidhts['DP1'] = 1.03
    barwidhts['DP2'] = 1.90

    # empy space in the detector (hole in the beam in cm)
    offset = {}
    offset['DP1'] = 7.5
    offset['DP2'] = 7.5
    
    # position in y  (estimate + .5)
    y = (chan + .5)*barwidhts[det] + offset[det]

    # position in x of the detectors
    xposition = {}
    xposition['DP1'] = 798
    xposition['DP2'] = 1471

    x = xposition[det]

    return x,y
```
